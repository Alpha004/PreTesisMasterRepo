## Analisis de Actividad Fisica.
Aunque es difícil cuantificar el «ejercicio» sin la intervención del usuario, podemos determinar cuándo los participantes están siendo más activos. Por lo tanto, desarrollamos un cálculo para los «periodos de actividad». Para calcular un «periodo de actividad», tomamos la magnitud media del vector de acelerometría y la frecuencia cardiaca media durante el intervalo de 5 minutos y lo comparamos con la media de los datos históricos anteriores del individuo. Si tanto el valor medio de la acelerometría como el de la frecuencia cardiaca en el intervalo de 5 minutos superan la media anterior, se considera que ese intervalo es un «periodo de actividad». 

```python
import pandas as pd

# Read the two CSV files
ACC = pd.read_csv('ACC_5min.csv')
HR = pd.read_csv('HR_5min.csv')
```

### Combinamos ambos datasets

```python
# Combine the DataFrames

# combined_df = pd.concat([ACC, HR], ignore_index=True)
combined_df = pd.merge(ACC, HR, on='datetime')

# Optionally, save the combined DataFrame to a new CSV file
combined_df_cleaned = combined_df.dropna(subset=['mean_x','mean_y'])
```

### Ahora empezaremos a calcular la actividad fisica

```python
# Inicializar la columna 'per_act' con ceros
combined_df_cleaned['per_act'] = 0

# Iterar sobre el DataFrame a partir de la segunda fila
for i in range(1, len(combined_df_cleaned)):
    if combined_df_cleaned['mean_x'].iloc[i] > combined_df_cleaned['mean_x'].iloc[i - 1] and combined_df_cleaned['mean_y'].iloc[i] > combined_df_cleaned['mean_y'].iloc[i - 1]:
        combined_df_cleaned['per_act'].iloc[i] = 1
```
