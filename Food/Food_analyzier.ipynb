### Trabajando con Datetime
Lo primero sera convertir los datetime a el formato correcto, ya que lo esta detectando como object, lo siguiente sera colocar como index las fechas y al final agrupar por cada 2, 8 y 24 horas los datos para obtener la sumatoria y media


```python
# Convertimos en fechas los datimetimes
food_values['time_begin'] = pd.to_datetime(food_values['time_begin'])
```

```python
food_values.head()
```

```python
food_values.info()
```

```python


## Se coloca indices como datetime
food_values = food_values.set_index('time_begin')
print(food_values.columns)
```

```python
# Empezamos por agrupar los valores importantes
# 2 horas
df_procesado_calorias_2h = food_values['calorie'].resample('2h') 
df_procesado_proteinas_2h = food_values['protein'].resample('2h') 
df_procesado_carbohidratos_2h = food_values['total_carb'].resample('2h') 
df_procesado_azucar_2h = food_values['sugar'].resample('2h')
# 8 horas 
df_procesado_calorias_8h = food_values['calorie'].resample('8h') 
df_procesado_proteinas_8h = food_values['protein'].resample('8h') 
df_procesado_carbohidratos_8h = food_values['total_carb'].resample('8h') 
df_procesado_azucar_8h = food_values['sugar'].resample('8h')
# 24 horas 
df_procesado_calorias_24h = food_values['calorie'].resample('24h') 
df_procesado_proteinas_24h = food_values['protein'].resample('24h') 
df_procesado_carbohidratos_24h = food_values['total_carb'].resample('24h') 
df_procesado_azucar_24h = food_values['sugar'].resample('24h')
```
