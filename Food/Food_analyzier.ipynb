## ANALISIS DE DATOS DE COMIDA (FOOD)

El presente es para analizar los datos de las comidas, el cual es un registro manual por cada usuario

```python
# Importando Pandas y otras librerias
import pandas as pd
import numpy as np
```

```python
# Parametros para calcular los valores
INTERVALO_2H = '2h'
INTERVALO_8H = '8h'
INTERVALO_24H = '24h'
```

```python
PACIENTE = '001'
PATH_FOLDER = 'G:\\Dataset\\big-ideas-lab-glycemic-variability-and-wearable-device-data-1.1.2\\'+PACIENTE+'\\'
```

```python
# Leyendo el CSV
food_values = pd.read_csv(PATH_FOLDER + 'Food_Log_'+PACIENTE+'.csv', engine='python', na_values="not available")
```

```python
# Convertimos en fechas los datimetimes
food_values['time_begin'] = pd.to_datetime(food_values['time_begin'])
```

### REAJUSTAR LOS TIEMPOS
Como se puede apreciar, los datos de los registros de comida no son precisos. Por ejemplo, el primer registro de comida se realiz√≥ en Febrero y los demas datos de los medidores son en Julio, si contar que son dias distintos.

```python
flag_wrong_date = False
if(PACIENTE == '016' or PACIENTE == '015'):
    food_values['datetime'] = food_values['time_begin'].apply(lambda x: x.replace(month=7))
    flag_wrong_date = True
elif (PACIENTE == '007'):
    food_values['datetime'] = food_values['time_begin'].apply(lambda x: x.replace(month=3))
    flag_wrong_date = True
elif (PACIENTE == '013'):
    food_values['datetime'] = food_values['time_begin'].apply(lambda x: x.replace(month=5))
    flag_wrong_date = True
else:
    food_values['datetime'] = food_values['time_begin']   

food_values.head()
```

```python
if(flag_wrong_date):
    # Fecha inicial
    match PACIENTE:
        case '016':
            fecha_inicial = pd.Timestamp('2020-07-16')
        case '015':
            fecha_inicial = pd.Timestamp('2020-07-19')
        case '013':
            fecha_inicial = pd.Timestamp('2020-05-28')
        case '007':
            fecha_inicial = pd.Timestamp('2020-03-14')
    
    # Actualizar las fechas de forma correlativa
    # Obtener las fechas √∫nicas y asignar nuevas fechas correlativas
    fechas_unicas = food_values['datetime'].dt.date.unique()
    nuevas_fechas = [fecha_inicial + pd.Timedelta(days=i) for i in range(len(fechas_unicas))]

    # Crear un diccionario para mapear las fechas originales a las nuevas
    mapeo_fechas = dict(zip(fechas_unicas, nuevas_fechas))

    # Asignar la nueva fecha usando el mapeo
    food_values['nueva_fecha'] = food_values['datetime'].dt.date.map(mapeo_fechas)

    food_values['datetime'] =  pd.to_datetime(food_values['nueva_fecha']) + food_values['datetime'].dt.time.apply(lambda x: pd.Timedelta(hours=x.hour, minutes=x.minute, seconds=x.second))
food_values.head()
```

### Trabajando con Datetime
Lo primero sera convertir los datetime a el formato correcto, ya que lo esta detectando como object, lo siguiente sera colocar como index las fechas y al final agrupar por cada 2, 8 y 24 horas los datos para obtener la sumatoria y media


```python
# Definir una funci√≥n para calcular el promedio de registros
def calculate_average_count(df, freq):
    resampled = df.resample(freq, on='datetime').size()
    return resampled.mean()
```

```python
# Empezamos por agrupar los valores importantes
# 2 horas
df_procesado_2h = food_values.resample(INTERVALO_2H, on='datetime').agg({
    'calorie': ['sum', 'mean'],
    'total_carb': ['sum', 'mean'],
    'sugar': ['sum', 'mean'],
    'protein': ['sum', 'mean'],
})
# # 8 horas 
df_procesado_8h = food_values.resample(INTERVALO_8H, on='datetime').agg({
    'calorie': ['sum', 'mean'],
    'total_carb': ['sum', 'mean'],
    'sugar': ['sum', 'mean'],
    'protein': ['sum', 'mean'],
})
# # 24 horas 
df_procesado_24h = food_values.resample(INTERVALO_24H, on='datetime').agg({
    'calorie': ['sum', 'mean'],
    'total_carb': ['sum', 'mean'],
    'sugar': ['sum', 'mean'],
    'protein': ['sum', 'mean'],
})

df_procesado_24h
```

```python


# Agregar columnas para contar la cantidad de registros en cada intervalo de 2 horas
df_procesado_2h['cantidad_comidas'] = food_values.resample(INTERVALO_2H, on='datetime').size()
df_procesado_8h['cantidad_comidas'] = food_values.resample(INTERVALO_8H, on='datetime').size()
df_procesado_24h['cantidad_comidas'] = food_values.resample(INTERVALO_24H, on='datetime').size()
df_procesado_2h
```

```python
# Ahora agregamos la media
average_count_2h = calculate_average_count(food_values, INTERVALO_2H)
df_procesado_2h['media_comidas'] = average_count_2h

average_count_8h = calculate_average_count(food_values, INTERVALO_8H)
df_procesado_8h['media_comidas'] = average_count_8h

average_count_24h = calculate_average_count(food_values, INTERVALO_24H)
df_procesado_24h['media_comidas'] = average_count_24h

df_procesado_24h
```

```python
# Renombrar las columnas para mayor claridad
df_procesado_2h.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_procesado_2h.columns.values]
df_procesado_8h.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_procesado_8h.columns.values]
df_procesado_24h.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_procesado_24h.columns.values]

print(df_procesado_8h)
```
